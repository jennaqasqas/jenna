```python
## 3.2 2D NumPy Arrays
```

```python
matrix = [
[5, 12, 17, 9],
[8, 21, 14, 3],
[6, 13, 19, 11]
]
```

```python
print(np.shape(matrix))
```

```python
#3 rows, 4 columns
```

```python
arr = np.array(matrix)
print(arr[:,2])
print(arr[1,:])
```

```python
print(arr * 2)
```

```python
print("mean of each column:", np.mean(arr, axis=0))
```

```python
print("mean:", np.mean(arr))
```

```python
print("row max:", np.max(arr, axis = 1))
```

```python
print("column min:", np.min(arr, axis=0))
```

```python
flat_arr = arr.flatten()
```

```python
flat_arr
```

```python
flat_arr.shape
```

```python
#becomes a 1D array with 12 elements
```

```python
print("min index:", np.argmin(flat_arr))
print("max index:", np.argmax(flat_arr))
```

```python
min_coord = np.unravel_index(np.argmin(flat_arr), arr.shape)
max_coord = np.unravel_index(np.argmax(flat_arr), arr.shape)
```

```python
print("coordinate of min:", min_coord)
print("coordinate of max:", max_coord)
```

```python
greater_than_15 = arr[arr > 15]
print("greater than 15:", greater_than_15)
```

```python
sorted_rows = np.sort(arr, axis=1)
print("rows sorted:\n", sorted_rows)
```

```python
## 4: Prime Star Systems
```

```python
arr = np.array([
[2, 3, 5],
[14, 6, 8],
[11, 13, 17],
[7, 10, 13]
])
```

```python
def prime_entry(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def prime_rows(array):
    mask = np.array([[prime_entry(x) for x in row] for row in array])
    return array[np.any(mask, axis=1)]

print(prime_rows(arr))
```
