```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as const
```

```python
df = pd.read_csv("/Users/jennaqasqas/python_decal_sp26/course_assignments/homework9/solar_system.csv")
```

```python
df_t = df.set_index("Attribute").T
df_t.index.name = "Planet"
df_t.reset_index(inplace=True)
df_t.columns.name = None
```

```python
print("original shape of dataset:", df.shape)
print("shape of transposed dataset:", df_t.shape)
print("original number of rows:", df.shape[0])
print("number of rows after transpose:", df_t.shape[0])
print("original number of columns:", df.shape[1])
print("number of columns after transpose:", df_t.shape[1])
```

df.set_index("Attribute") makes the "attribute" column vecome the row labels. .T transposes the dataset, so the rows and columns are flipped.

df.index.name = "Planet" names the index "Planet", and df.reset_index(inplace=True) turns that index into a real column. df.columns.name = None removes the extra label above the columns so the table looks cleaner.

The new shape is not (11, 20) because "Attribute" stopped being a normal column when it became the index, and then "Planet" was added back as a new column after reset_index(). So the structure was also changed.

## 2.1.2 Investigate Data Types

```python
df_t.dtypes
```

```python
for col in df_t.columns:
    print(f"\ncolumn: {col}")
    print(df_t[col].apply(type).value_counts())
```

a) The actual values are strings. Even values that look like numbers are stored as text rather than numerical data.

b) This is problematic because we can't perform mathematical operations on strings... can't extract statistical data like averages or make plots.

## 2.1.3 Convert Strings to Numbers

```python
for col in df_t.columns:
    if col not in ["Planet", "Ring System?", "Global Magnetic Field?"]:
        df_t[col] = pd.to_numeric(df_t[col], errors="coerce")

print(df_t.dtypes)
```

a) The three datatypes are str, float64, and int64.

b) str means text values like words or labels. float64 means decimal numbers. int64 means the data is a whole number

c) The columns still stores as strings are "Planet," "Ring System?," and "Global Magnetic Field" because these are text values, not nuermical. Planet has planet names and the others are yes/no.

```python
def attach_units(df, old_col, new_col, unit):
    new_values = []
    for val in df[old_col]:
        new_values.append(val * unit)
    
    df[old_col] = new_values
    df.rename(columns={old_col: new_col}, inplace=True)
```

```python
attach_units(df_t, "Mass (10^24kg)", "Mass (kg)", 1e24 * u.kg)

print(df_t["Mass (kg)"])
```

## 2.3 Data Analysis

```python
df_t["Perihelion (10^6 km)"] = df_t["Perihelion (10^6 km)"] * (1e6 * u.km)
df_t["Aphelion (10^6 km)"] = df_t["Aphelion (10^6 km)"] * (1e6 * u.km)

semi_major_axis = (df_t["Perihelion (10^6 km)"] + df_t["Aphelion (10^6 km)"]) / 2

aphelion_index = df_t.columns.get_loc("Aphelion (10^6 km)")
df_t.insert(aphelion_index + 1, "Semi-Major Axis (km)", semi_major_axis)

print("columns:")
print(df_t.columns)

print("\nsemi-major axis:")
print(df_t["Semi-Major Axis (km)"])
```

```python
df_t["Orbital Period (days)"] = df_t["Orbital Period (days)"].apply(lambda x: (x * u.day).to(u.yr))
df_t.rename(columns={"Orbital Period (days)": "Orbital Period (yr)"}, inplace=True)

print("\ncolumns after conversion:")
print(df_t.columns)

print("\nOrbital Period (yr):")
print(df_t["Orbital Period (yr)"])
```

```python
earth = df_t[df_t["Planet"] == "Earth"].iloc[0]
print(f"earth orbital period: {earth['Orbital Period (yr)'].value:.4f} {earth['Orbital Period (yr)'].unit}")
```

An AU is the average distance between Earth and the Sun. 

```python
print("\nconst.au:")
print(const.au)
```

```python
print("\none AU in kilometers:")
print(const.au.to(u.km))
```

```python
df_t["Diameter (km)"] = df_t["Diameter (km)"].apply(lambda x: (x * u.km).to(u.au))
df_t.rename(columns={"Diameter (km)": "Diameter (AU)"}, inplace=True)

df_t["Distance from Sun (10^6 km)"] = df_t["Distance from Sun (10^6 km)"].apply(lambda x: (x * 1e6 * u.km).to(u.au))
df_t.rename(columns={"Distance from Sun (10^6 km)": "Distance from Sun (AU)"}, inplace=True)

df_t["Perihelion (10^6 km)"] = df_t["Perihelion (10^6 km)"].apply(lambda x: (x * 1e6 * u.km).to(u.au) if not hasattr(x, "unit") else x.to(u.au))
df_t.rename(columns={"Perihelion (10^6 km)": "Perihelion (AU)"}, inplace=True)

df_t["Aphelion (10^6 km)"] = df_t["Aphelion (10^6 km)"].apply(lambda x: (x * 1e6 * u.km).to(u.au) if not hasattr(x, "unit") else x.to(u.au))
df_t.rename(columns={"Aphelion (10^6 km)": "Aphelion (AU)"}, inplace=True)
```

```python
venus = df_t[df_t["Planet"] == "Venus"].iloc[0]

print("\nsample converted values for Venus:")
print("diameter (AU):", venus["Diameter (AU)"])
print("distance from Sun (AU):", venus["Distance from Sun (AU)"])
print("perihelion (AU):", venus["Perihelion (AU)"])
print("aphelion (AU):", venus["Aphelion (AU)"])
```
