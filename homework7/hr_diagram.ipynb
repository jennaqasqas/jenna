```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("/Users/jennaqasqas/python_decal_sp26/course_assignments/homework7/star_data.txt",
skiprows=1)
```

```python
data.shape
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
print(np.min(magnitude))
print(np.max(magnitude))

brightest_star_index = np.argmin(magnitude)
dimmest_star_index = np.argmax(magnitude)

print(data[brightest_star_index])
print(data[dimmest_star_index])
```

```python
print(np.min(temperature))
print(np.max(temperature))

coldest_star_index = np.argmin(temperature)
hottest_star_index = np.argmax(temperature)

print(data[coldest_star_index])
print(data[hottest_star_index])
```

```python
print("Brightest star:", data[np.argmin(magnitude)])
print("Dimmest star:", data[np.argmax(magnitude)])
```

```python
mag_log = np.log(magnitude)
temp_log = np.log(temperature)

plt.figure(figsize=(8, 6))
plt.scatter(temp_log, mag_log, label = "Stars", c=temperature, cmap = "coolwarm")

plt.scatter(np.log(temperature[brightest_star_index]),
            np.log(magnitude[brightest_star_index]),
            color="black", marker="*", s=300, label="Brightest star")

plt.scatter(np.log(temperature[dimmest_star_index]),
            np.log(magnitude[dimmest_star_index]),
            color="purple", marker="*", s=300, label="Dimmest star")

plt.scatter(np.log(temperature[hottest_star_index]),
            np.log(magnitude[hottest_star_index]),
            color="green", marker="*", s=300, label="Hottest star")

plt.scatter(np.log(temperature[coldest_star_index]),
            np.log(magnitude[coldest_star_index]),
            color="orange", marker="*", s=300, label="Coldest star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.title("Hertzsprung-Russel Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.legend()


plt.colorbar(label="Temperature (K)")

plt.show()
```
