```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
for A in amplitude:
    y = make_sine_wave(x, A, w=1)
    plt.plot(x, y, label=f"A = {A}")

plt.title("Sine Waves — Varying Amplitude (ω = 1)")
plt.xlabel("x")
plt.ylabel("y = A · sin(ωx)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(True, alpha=0.3)

# --- Plot 2: Varying Frequency (fixed amplitude A=1) ---
plt.subplot(1, 2, 2)
for w in frequency:
    y = make_sine_wave(x, A=1, w=w)
    plt.plot(x, y, label=f"ω = {w}")

plt.title("Sine Waves — Varying Frequency (A = 1)")
plt.xlabel("x")
plt.ylabel("y = A · sin(ωx)")
plt.axhline(0, color='black', linewidth=0.8, linestyle='--')
plt.legend()
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("sine_waves.png", dpi=150)
plt.show()

print("Done! Plot saved as sine_waves.png")
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)

amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]

fig, axes = plt.subplots(1, 3, figsize=(25, 5))

for A, w in zip(amplitude, frequency):

    y1 = make_sine_wave(x, A=A, w=1)
    axes[0].plot(x, y1, label=f"A = {A}", linewidth=2)

    y2 = make_sine_wave(x, A=1, w=w)
    axes[1].plot(x, y2, label=f"ω = {w}", linewidth=2)

    y3 = make_sine_wave(x, A=A, w=w)
    axes[2].plot(x, y3, label=f"A={A}, ω={w}", linewidth=2)

font = {"fontsize": 14}

axes[0].set_title("Varying Amplitude (ω = 1)", **font, fontweight="bold")
axes[0].set_xlabel("x", **font)
axes[0].set_ylabel("y = A · sin(x)", **font)
axes[0].legend(fontsize=12)
axes[0].grid(True, alpha=0.3)
axes[0].axhline(0, color="black", linewidth=0.8, linestyle="--")

axes[1].set_title("Varying Frequency (A = 1)", **font, fontweight="bold")
axes[1].set_xlabel("x", **font)
axes[1].set_ylabel("y = sin(ωx)", **font)
axes[1].legend(fontsize=12)
axes[1].grid(True, alpha=0.3)
axes[1].axhline(0, color="black", linewidth=0.8, linestyle="--")

axes[2].set_title("Varying Amplitude & Frequency", **font, fontweight="bold")
axes[2].set_xlabel("x", **font)
axes[2].set_ylabel("y = A · sin(ωx)", **font)
axes[2].legend(fontsize=12)
axes[2].grid(True, alpha=0.3)
axes[2].axhline(0, color="black", linewidth=0.8, linestyle="--")

plt.tight_layout()

plt.show()
```
